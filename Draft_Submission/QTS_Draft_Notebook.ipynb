# 1.1) Motivation and Hypothesis

### Why On-Chain data?

Why On-Chain data?

- On-chain data is central to cryptocurrency price prediction because it reflects core activity on the blockchain itself, capturing how   participants actually use, trade, and move coins. For example, on-chain variables—like the volume of transactions, distribution of addresses, mining difficulty, and total number of active addresses—can illuminate real-time shifts in network fundamentals that do not appear in conventional price or market data alone. As Kim et al. note, these blockchain-derived factors are distinct from those used for traditional assets and therefore require new approaches to accurately model cryptocurrency prices. By incorporating on-chain data into their LSTM-based framework, they achieved an MAE of 0.3462 in predicting Bitcoin’s price, which illustrates how directly modeling blockchain-level activity can yield more precise and robust forecasts than relying solely on historical price or external economic indicators.


### Why change in point detection to segment the data?

-   Since the price moves in an unexpected range that has been previously unseen, constructed
    machine learning-based models are not able to predict future
    spot prices accurately. This problem does not apply only to cer-
    tain prediction algorithms but could affect practically every
    prediction model constructed based on price data within a
    moderate range. This work therefore proposes a novel method
    to address the aforementioned problem using a change point detection (CPD) technique.


### Why techinical data?

Technical analysis is crucial for price prediction models. Parente et al., for instance, leveraged OHLCV data to generate various technical indicators—such as Bollinger Bands, EMAs, and RSI—and developed an MLP-based model that achieved positive returns across differing economic climates.

### Prediction Framework:

The proposed framework consists of five phases.

    1.⁠ ⁠Feature sets have been collected from on-chain data and price history data.
    2.⁠ ⁠The segmentation of input data based on the CPD technique (PELT) is conducted.
    3.⁠ ⁠The proposed price prediction model, utilizing 3-class (Buy, Hold, Sell) MLP. 
    4.⁠ ⁠Portfolio allocation is predicated on the "accuracy score" of the model's predictions
    5.⁠ ⁠Risk management includes utilizing an exponetial decay method for repeated orders on the same token
<h5>Our model bridges this gap by combining both data types and various methodologies from  Parente et al. and Kim et al

# 1.2) Proof of Work (PoW) vs. Proof of Stake (PoS) and On-Chain Metrics (More on the Hypothesis)

---

## Proof of Work (PoW)

### Mechanism
- **Mining Process:**  
  In PoW systems (e.g., Bitcoin), miners compete to solve complex cryptographic puzzles. They do this by trying random cryptographic hashes until they find the matching one necessary to verify and process the block.  
- **Block Creation:**  
  The first miner to solve the puzzle creates the next block and is rewarded in the digital currency associated with the blockchain. 
- **Resource Use:**  
  The probability of solving the puzzle is directly proportional to the computational power (hash rate) a miner dedicates to the network.

### Energy Consumption & Price Relationship
- **High Energy Demand:**  
  Solving puzzles requires massive computational power, leading to significant energy consumption.
- **On-Chain Metrics:**  
  Metrics such as hash rate and block difficulty are proxies for the energy and hardware investment required.
- **Price Floor:**  
  Since miners must cover their costs (electricity, hardware depreciation), these energy-related metrics tend to be **contemporaneously linked** to the coin's market price. An increase in energy cost or hash rate may immediately impact price.

---

## Proof of Stake (PoS)

### Mechanism
- **Validator Selection:**  
  In PoS systems, validators are chosen based on the number of coins they “stake” (lock up) as collateral. If a validator verifies a bad block, they will incurr a slash penalty, which takes away a proportion of the coins they staked. This causes built-in trust in the validators, because they generally have the most to lose.
- **Block Creation:**  
  The more coins you're willing to stake, the more likely you're going to be chosen by the system to be a validator.
- **Less Energy Intensive:**  
  Unlike PoW, PoS doesn’t require high computational power, shifting the cost from energy to capital allocation. For example, when Ethereum switched from PoW to PoS, energy consumption on the blockchain reduced 99%.

### Staking and Price Relationship
- **Market Confidence:**  
  The amount of staked tokens reflects the validators’ commitment and market confidence in the network.
- **On-Chain Metrics:**  
  Gas limit, gas used, block size, difficulty, and staking behavior
- **Lead-Lag Effect:**  
  Changes in staking behavior might **lead** future spot price movements rather than being immediately reflected in the price, due to the nature of long-term capital allocation decisions.

---

## Rationale Behind the Hypotheses

### On-Chain Metrics and PoW Coin Prices
**Hypothesis:**  
_On-chain data metrics are fundamentally related to the price of PoW coins due to the nature of its block processing system._

**Rationale:**  
- **Cost of Production:**  
  Mining incurs real-world costs (electricity, hardware), which are captured by metrics like hash rate and difficulty.  
- **Immediate Impact:**  
  Since miners operate based on covering these costs, any change in these metrics is rapidly incorporated into the coin’s price, and traders are directly looking at these analytics when trading.

---

### Energy Consumption’s Role in PoW Metrics
**Hypothesis:**  
_Processing blocks in PoW systems is fundamentally related to energy consumption, making these metrics more useful in explaining the price of the digital asset._

**Rationale:**  
- **Energy as a Fundamental Input:**  
  The work required to mine (and thus the energy consumed) is a direct input in the network's security.
- **Cost-Price Link:**  
  Variations in energy costs affect mining profitability, setting a lower bound on the coin’s price.

---

### PoS Metrics and Validator Behavior
**Hypothesis:**  
_Processing blocks in PoS systems is fundamentally related to how much validators are staking._

**Rationale:**  
- **Staking and Processing Difficulty as a Signal:**  
  The total amount staked reflects the level of market confidence and reduces the circulating supply. Additionally, processing difficulty can still give us useful information about the energy consumption of the network, which is indirectly related to staking behavior.
- **Delayed Reflection:**  
  Since staking is a strategic, long-term decision, changes in these metrics may signal future price moves rather than moving in lockstep with the price. Also, since energy consumption is not as popular of a metric in PoS coins, we believe this is our edge.

---

### Predictive Power: PoW vs. PoS Metrics

**For PoW:**

**Hypothesis:**  
_Because PoW on-chain metrics have a fundamental, contemporaneous relationship to asset price, there wouldn't be as much predictive power looking into the future._

**Rationale:**  
- **Immediate Reflection:**  
  The mining cost is immediately factored into the coin’s price, leaving little lag for prediction.
- **Limited Forecasting:**  
  These metrics reflect current conditions and are less useful as leading indicators.

**For PoS:**

**Hypothesis:**  
_PoS on-chain metrics have a lead-lag relationship to digital asset price, where the information contained in the on-chain metrics isn't instantaneously reflected in digital asset price, but instead, it's leading the asset price._

**Rationale:**  
- **Capital Reallocation Delay:**  
  Validators’ staking decisions are made over a longer time horizon, creating a delay between changes in on-chain metrics and price adjustments.
- **Predictive Window:**  
  This delay offers a window during which changes in staking behavior can be used to predict future price movements.

**Hypothesis:**  
_As a result, this lead-lag relationship can leave us a window to predict price directions in a profitable quantitative trading strategy._

**Rationale:**  
- **Trading Opportunities:**  
  Early signals from increased staking can be exploited to position ahead of anticipated price moves.
- **Time-Lag Exploitation:**  
  The delay between on-chain metric changes and market reaction provides a basis for developing predictive models and profitable trading strategies.

# 1.3) Model Construction

### We’ve chosen to build our model using the MLP base proposed by Parente et. al. The model itself is a 4-layer Multi-layer Percentron (MLP) model.

- The first layer contains all the data discussed in the feature space. This includes the price data for the token, technical indicators built off of the closing prices (e.g. VWAP), and features derived from on-chain data (e.g. block difficulty).

- The two hidden layers will feature a number of nodes to be determined throughout the construction of the model, as we feel this is better approximated once we have tested feeding data through the model. As was done in the base paper, we will be employing the LeakyReLU activation function. This activation function is similar to the standard ReLU function, except that negative values are returned at a deflated value.

- The output layer will utilize the weights between the second hidden layer and its own layer. Per the paper we will utilize the Softmax function across the 3 nodes for its repetitive action (buy/sell/hold). The sum of the three probabilities will equal 1, and the highest probability value will be the recommended trading action. The portfolio allocation will be based on the magnitude of the Softmax prediction.

- The model itself features the following parameters:​

    1. $b :=$ lookback window size​

    2. $f :=$ prediction horizon​

    3. $x :=$ lower prediction threshold​

    4. $y :=$ upper prediction threshold

```python
import requests
import pandas as pd
import time
from dotenv import load_dotenv


import os
load_dotenv()
# Your CoinAPI API Key
API_KEY= os.getenv("coinapi")
def get_crypto_data(
    crypto="BTC",
    quote_currency="USDT",
    exchange="BINANCE",
    start_date="2019-02-25T00:00:00",
    end_date="2025-02-25T00:00:00",
    period_id="1DAY"
):

    # Construct the symbol ID based on the parameters
    symbol = f"{exchange}_SPOT_{crypto}_{quote_currency}"

    url = f"https://rest.coinapi.io/v1/ohlcv/{symbol}/history"
    headers = {"X-CoinAPI-Key": API_KEY}


    params = {
        "period_id": period_id,
        "time_start": start_date,
        "time_end": end_date,
        "limit": 10000  # Max records per request
    }

    all_data = []
    max_retries = 5

    # print(f"Fetching {crypto}/{quote_currency} data from {exchange} with {period_id} candles")
    # print(f"Symbol ID: {symbol}")
    # print(f"Date range: {start_date} to {end_date}")

    while True:
        retry_count = 0
        response = None

        # Retry loop for handling rate limits
        while retry_count < max_retries:
            response = requests.get(url, headers=headers, params=params)

            # If we get a rate limit error (429)
            if response.status_code == 429:
                # Try to get the retry-after value
                retry_after = 15  # Default to 15 seconds

                try:
                    # Try to parse JSON response for retry-after info
                    error_data = response.json()
                    if "Retry-After" in error_data:
                        retry_after = int(error_data["Retry-After"])
                except:
                    # If can't parse JSON, try to get from headers
                    if "Retry-After" in response.headers:
                        retry_after = int(response.headers["Retry-After"])

                # print(f"Rate limit hit. Waiting {retry_after} seconds before retrying...")
                time.sleep(retry_after + 1)  # Add 1 second buffer
                retry_count += 1
                continue

            # If we get any other error or success, break out of retry loop
            break

        # If we've exhausted our retries and still have rate limit issue
        if retry_count >= max_retries and response.status_code == 429:
            # print(f"Still hitting rate limits after {max_retries} retries. Aborting.")
            break

        # Regular error handling
        if response.status_code != 200:
            print(f"Error: Status code {response.status_code}")
            print(f"Response: {response.text}")

            # If symbol not found, try alternative quote currencies
            if response.status_code == 404 or (response.text and "invalid or does not exist" in response.text):
                return None
            break

        # JSON parsing
        try:
            if response.text:
                data = response.json()
            else:
                print("Empty response received")
                break
        except Exception as e:
            print(f"JSON parsing error: {e}")
            print(f"Response content: {response.text[:200]}...")
            break

        # API error checking
        if isinstance(data, dict) and "error" in data:
            print("Error:", data["error"])
            break

        # Empty data check
        if not data:
            print("No more data available")
            break

        # Successfully got data
        df = pd.DataFrame(data)
        all_data.append(df)

        print(f"Retrieved {len(df)} records")

        # Update for next batch
        if len(df) > 0:
            last_timestamp = df["time_period_end"].iloc[-1]
            params["time_start"] = last_timestamp
            # Wait between successful requests too, to avoid hitting limits
            time.sleep(5)
        else:
            break

    # Process results
    if all_data:
        full_df = pd.concat(all_data, ignore_index=True)
        full_df["time_period_start"] = pd.to_datetime(full_df["time_period_start"])
        full_df.set_index("time_period_start", inplace=True)
        return full_df
    else:
        return None


def fetch_with_fallbacks(
        crypto="BTC",
        start_date="2019-02-25T00:00:00",
        end_date="2025-02-25T00:00:00",
        period_id="1DAY"
):

    # Try different quote currencies on Binance first
    quote_currencies = ["USDT", "USD", "BUSD"]
    for quote in quote_currencies:
        df = get_crypto_data(
            crypto=crypto,
            quote_currency=quote,
            exchange="BINANCE",
            start_date=start_date,
            end_date=end_date,
            period_id=period_id
        )

        if df is not None and not df.empty:
            print(f"Success with {crypto}/{quote} on BINANCE!")
            print(f"Total records: {len(df)}")
            print(f"Date range: {df.index.min()} to {df.index.max()}")
            return df

    # If Binance failed, try other major exchanges
    exchanges = ["COINBASE", "KRAKEN", "BITSTAMP", "KUCOIN"]
    for exchange in exchanges:
        for quote in quote_currencies:
            df = get_crypto_data(
                crypto=crypto,
                quote_currency=quote,
                exchange=exchange,
                start_date=start_date,
                end_date=end_date,
                period_id=period_id
            )

            if df is not None and not df.empty:
                print(f"Success with {crypto}/{quote} on {exchange}!")
                print(f"Total records: {len(df)}")
                print(f"Date range: {df.index.min()} to {df.index.max()}")
                return df

    print(f"Failed to retrieve data for {crypto} from any exchange or quote currency.")

    return None
```

# 3.) Data Cleaning

```python
import pandas as pd
import numpy as np

def clean_data(data, price_data=None, on_chain_data=None, set_index=True):
    if price_data and on_chain_data:
        raise ValueError('Invalid Parameter Values: Both price_data or on_chain data cannot be True')
    elif price_data:
        prices = data.copy()
        if prices.isna().any().any():
            print('you have nans here')
            return prices 
        
        if set_index:
            prices['time_period_end'] = pd.to_datetime(prices['time_period_end'])
            prices = prices.set_index('time_period_end') 
        else:
            prices.index = pd.to_datetime(prices.index)

        prices['time_open'] = pd.to_datetime(prices['time_open'])
        prices['time_close'] = pd.to_datetime(prices['time_close'])
        return prices
    
    elif on_chain_data:
        metrics = data.copy()
        if metrics.isna().any().any():
            print('you have nans here')
            return metrics 
        
        if set_index:
            metrics['hour'] = pd.to_datetime(metrics['hour'])
            metrics = metrics.set_index('hour')
        else:
            metrics.index = pd.to_datetime(metrics.index)
        return metrics
    else:
        raise ValueError('Invalid Parameter Values: price_data or on_chain data must be True')

def check_missing_hours(df):
    """
    Check if the DataFrame's datetime index skips any hourly datapoints.
    
    Parameters:
        df (pd.DataFrame): DataFrame with a DatetimeIndex.
        
    Returns:
        missing (pd.DatetimeIndex): The missing hourly timestamps.
    """
    # Ensure the index is a DatetimeIndex
    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("DataFrame index must be a DatetimeIndex.")
    
    # Create an expected date_range from the minimum to the maximum timestamp at hourly frequency
    expected_range = pd.date_range(start=df.index.min(), end=df.index.max(), freq='H')
    
    # Determine which timestamps are missing
    missing = expected_range.difference(df.index)
    return missing


def preprocess_data(data, price_data=None, on_chain_data=None, set_index=True):
    if price_data and on_chain_data:
        raise ValueError('Invalid Parameter Values. Both price_data and on_chain_data cannot both be True')

    elif (not price_data) and (not on_chain_data):
        raise ValueError('Invalid Parameter Values. Both price_data and on_chain_data cannot both be False')
    
    else:
        # Create the target df and merge on correct dates. Then forward fill the na values
        if price_data:
            df = clean_data(data.copy(), price_data=True, set_index=True) if set_index else clean_data(data.copy(), price_data=True, set_index=False)
            target_df = pd.DataFrame(0, columns=[0], index=pd.date_range(start=df.index.min(), end=df.index.max(), freq='H'))
        
        elif on_chain_data:
            df = clean_data(data.copy(), on_chain_data=True, set_index=True) if set_index else clean_data(data.copy(), on_chain_data=True, set_index=False)
            target_df = pd.DataFrame(0, columns=[0], index=pd.date_range(start=df.index.min(), end=df.index.max(), freq='H'))

        target_df = target_df.join(df, how='left').drop(0, axis=1)
        target_df = target_df.fillna(method='ffill')


        return target_df
```

```python
def plot_cumulative_returns(
    df, 
    time_col='time_close', 
    price_col='price_close', 
    title='Cumulative Returns Over Time'
):
    
    df = df.sort_values(by=time_col).reset_index(drop=True)
    
    df['returns'] = df[price_col].pct_change()
    
    df['cumulative_returns'] = (1 + df['returns']).cumprod() - 1
    
    plt.figure(figsize=(10, 6))
    plt.plot(df[time_col], df['cumulative_returns'], label='Cumulative Returns')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Cumulative Returns')
    plt.legend()
    plt.grid(True)
    plt.show()
```
